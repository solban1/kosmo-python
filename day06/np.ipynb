```python
import numpy as np

# standard deviation
a = [87, 88, 89, 87, 88, 86, 87]
b = [33, 112, 139, 29, 60, 78, 98]

print(np.std(a))
print(np.var(a))
print(np.std(b))
print(np.var(b))
```
