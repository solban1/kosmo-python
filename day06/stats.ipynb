```python
import matplotlib.pyplot as plt
from scipy import stats

x = [6, 8, 9, 8, 3, 18, 3, 10, 5, 12, 13, 10, 7]
y = [100, 87, 88, 89, 112, 87, 104, 88, 95, 79, 78, 86, 87]
plt.scatter(x, y)
# plt.show()

slope, intercept, rvalue, pvalue, stderr = stats.linregress(x, y, "less")

print("slope", slope)  # 기울기, 가중치(weight, 영향력?)
print("intercept", intercept)  # y절편, 편향(bias)

# Pearson correlation coefficient: 점과 선의 상관관계 지수
# -1에 가까울수록 반비례(음의 상관관계), 1에 가까울수록 비례(양의 상관관계)
print("rvalue", rvalue)

# p값(0~1), 극단적인 결과가 관측될 확률
# 0에 가까울수록(<.05) xy간의 상관관계가 높다고 함
print("pvalue", pvalue)

print("stderr", stderr)


def f(x):
    return slope * x + intercept


model = list(map(f, x))  # the new y
plt.plot(x, model, "r-")
plt.show()

# predict y from x
print("f(16):", f(16))
```
